```python
df_2=df_1.drop(['_id','timezone_offset','minutely','current.weather','current.snow.1h','current.rain.1h','alerts'],axis=1)
```

```python
df_2.head()
```

```python
from matplotlib import pyplot as plt
# Latitude vs. Temperature Plot
plt.scatter(df_2["lat"], df_2["current.temp"], marker="o", s=15, color='blue')
plt.title("City Latitute vs  Temperature")
plt.ylabel("Temperature(F)")
plt.xlabel("Latitude")
plt.grid(True)
plt.show()
```

```python
# Latitude vs. Humidity Plot
plt.scatter(df_2["lat"], df_2["current.humidity"], marker="o", s=15, color='blue')
plt.title("City Latitute vs Humidity")
plt.ylabel("Humidity(%)")
plt.xlabel("Latitude")
plt.grid(True)
plt.show()
```

```python
# Latitude vs. Cloudiness Plot
plt.scatter(df_2["lat"], df_2["current.clouds"], marker="o", s=15, color='blue')
plt.title("City Latitute vs Cloudiness")
plt.ylabel("Cloudiness(%)")
plt.xlabel("Latitude")
plt.grid(True)
plt.show()
```

```python
#Latitude vs. Wind Speed Plot
plt.scatter(df_2["lat"], df_2["current.wind_speed"], marker="o", s=15, color='blue')
plt.title("City Latitute vs Wind Speed")
plt.ylabel("Wind Speed(mph)")
plt.xlabel("Latitude")
plt.grid(True)
plt.show()
```
